```python
import numpy as np
import matplotlib.pyplot as plt
```

Veamos el ejemplo de escribir $f(x)=1$ definida en el intervalo $\left [0, L \right ]$ como suma de senos. 

En ese caso vimos que se puede escribir $$ f(x) = 1 = \sum_{p = 0 }^{\infty}\frac{4}{\pi(2p+1)}\sin\left (\frac{\pi x}{L}(2p+1)\right )$$

```python
L = 1
numero_de_terminos = 1
x = np.linspace(0, L, 1000)
terminos = [4 / (np.pi * (2*p+1)) * np.sin(np.pi*x/L * (2*p+1)) for p in range(numero_de_terminos)]
f  = np.sum(terminos, axis = 0)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (12, 4))
for p in range(numero_de_terminos):
    ax1.plot(x, terminos[p], label = f'p = {p}')
ax1.legend();
ax2.plot(x, f, label = 'Reconstruccion');
ax2.plot(x, np.ones(len(x)), 'r', label = 'f(x) = 1)')
ax2.legend();
```

Veamos el ejemplo de escribir $f(x)=x$ definida en el intervalo $\left [0, L \right ]$ como suma de senos. 

En ese caso vimos que se puede escribir $$f(x) = x =  \sum_{m = 1 }^{\infty}\frac{2L}{m\pi}(-1)^{m+1}\sin\left (\frac{\pi x m}{L}\right )$$

```python
L = 1
numero_de_terminos = 1
x = np.linspace(0, L, 1000)
terminos = [(2*L) / (np.pi * p) * (-1)**(p+1) * np.sin(np.pi*x/L * p) for p in range(1, numero_de_terminos+1)]
f  = np.sum(terminos, axis = 0)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (12, 4))
for p in range(1, numero_de_terminos+1):
    ax1.plot(x, terminos[p-1], label = f'p = {p}')
ax1.legend();
ax2.plot(x, f, label = 'Reconstruccion');
ax2.plot(x, x, 'r', label = 'f(x) = 1)')
ax2.legend();
```
