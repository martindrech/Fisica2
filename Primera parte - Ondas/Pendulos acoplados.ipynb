Lo primero que vamos a hacer es importar librerias que nos van a servir: numpy, para hacer calculos, y pyplot de matplotlib para hacer graficos. 

```python
import numpy as np
import matplotlib.pyplot as plt
```

Bueno, ahora definamos los parametros del problema. Tenemos las masas $m_a$ y $m_b$, las constantes $g$ y $k$ y el largo de los hilos $l$. De paso tambien definimos las frecuencias normales. 

```python
ma = 1
mb = 1.5
g = 10
l = 1
k = 1
w1 = np.sqrt(g/l)
w2 = np.sqrt(g/l + k/ma + k/mb)
M = ma+mb
```

Bueno, puedo escribir las soluciones que encontramos:
$$
\psi_a(t) = \frac{m_b}{M}\left [ \cos(\omega_1 t)-\cos(\omega_2 t) \right ]\\
\psi_b(t) = \frac{m_b}{M}\left [ \cos(\omega_1 t)+\frac{m_a}{m_b}\cos(\omega_2 t) \right ]
$$


```python
t = np.linspace(0, 50, 1000)
psi_a = mb/M * (np.cos(w1*t) - np.cos(w2*t))
psi_b = mb/M * (np.cos(w1*t) + ma/mb * np.cos(w2*t))
```

```python
plt.plot(t, psi_a, label = '$\psi_a$');
plt.plot(t, psi_b, label = '$\psi_b$');
plt.legend()
```

Hagamos lo mismo con las energias cineticas medias: 
$$
\left \langle T_a \right \rangle = \frac{m_a m_b^2}{M^2} ~ \bar{\omega}^2 \sin^2(\Delta \omega ~  t) \\
\left \langle T_b \right \rangle = \frac{m_b}{4} ~ \bar{\omega}^2 \left [ \cos^2(\Delta \omega ~t)+(\frac{m_a-m_b}{m_a+m_b})^2\sin^2(\Delta \omega ~t) \right ]
$$

```python
wm = (w1+w2)/2
dw = (w2-w1)/2
Ta = ma*mb**2/M**2 * wm**2 * np.sin(dw*t)**2
Tb = mb/4 *wm**2 * (np.cos(dw*t)**2  + ((ma-mb)/(ma+mb))**2 * np.cos(dw*t)**2)
```

```python
plt.plot(t, Ta, label = 'Ta')
plt.plot(t, Tb, label = 'Tb')
plt.plot(t, Ta + Tb, label = 'Ta + Tb')
plt.legend();
```
