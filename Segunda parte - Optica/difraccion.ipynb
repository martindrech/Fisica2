```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def Intensidad(theta, theta0, a, b, N, wl):
  k = 2*np.pi/wl
  alpha = a*k/2 * (np.sin(theta)-np.sin(theta0))
  beta = b*k/2 * (np.sin(theta)-np.sin(theta0))
  return N**-2 * np.sinc(beta)**2 * np.sin(N*alpha)**2/np.sin(alpha)**2, np.sinc(beta)**2                        
```

```python
y = np.linspace(-5, 5, 1000)*1e-3
L = 3
thetas = np.arctan2(y,L)
theta0 = 0
a = 1e-3
b = 50e-6
N = 20
wl = 700e-9
dwl = 10e-9
I, difr = Intensidad(thetas, theta0, a, b, N, wl) 
I2, difr2 = Intensidad(thetas, theta0, a, b, N, wl+dwl)
plt.figure(figsize = (12, 6))
plt.plot(y*1e3, I, 'b')
plt.plot(y*1e3, I2, 'r')
plt.plot(y*1e3, difr, '--')
```
